#### 4 Calculate state-wise:

```python
df_copy.groupby(['State_name']).Population.sum().sort_values(ascending=False)
```

```python
df_copy.groupby(['State_name'])['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains'].sum()
```
